```python
from bs4 import BeautifulSoup
import requests

# action_url = 'https://www.rottentomatoes.com/top/bestofrt/top_100_comedy_movies/'
# a_page = requests.get(action_url)

# soup = BeautifulSoup(a_page.content, 'html.parser')
# # soup.prettify
# movie_title = soup.find_all('a', class_='unstyled articleLink')

genres =['action__adventure', 'comedy', 'horror', 'romance']
for i in genres:
    url = 'https://www.rottentomatoes.com/top/bestofrt/top_100_{}_movies/'.format(i)
    url_page = requests.get(url)
    soup = BeautifulSoup(url_page.content, 'html.parser')
    movie_title = soup.find_all('a', class_='unstyled articleLink')
    for movie in movie_title:
        print(movie.text)
```
